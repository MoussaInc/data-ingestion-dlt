```python
import duckdb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)
sns.set_style('whitegrid')
```

```python
# Connexion à la base
con = duckdb.connect('open_library.duckdb')
```

```python
# Lister les schémas
schemas = con.execute("SHOW SCHEMAS").fetchdf()
print("📁 Schémas disponibles:")
print(schemas)
```

```python
# Structure de la table books
structure = con.execute("DESCRIBE library.books").fetchdf()
print("\n🏗️  Structure de la table 'books':")
print(structure)
```

```python
# Lister toutes les tables du schéma library
tables = con.execute("SHOW TABLES FROM library").fetchall()
table_names = [t[0] for t in tables]

print(f"📊 {len(table_names)} tables trouvées dans le schéma 'library':\n")

# Afficher la structure de chaque table
for table_name in table_names:
    print(f"\n{'='*60}")
    print(f"📋 Table: {table_name}")
    print('='*60)
    
    structure = con.execute(f"DESCRIBE library.{table_name}").fetchdf()
    print(structure.to_string(index=False))
    
    # Compter les lignes
    count = con.execute(f"SELECT COUNT(*) FROM library.{table_name}").fetchone()[0]
    print(f"\n📊 Nombre de lignes: {count:,}")
```


## 📈 Statistiques globales


```python
# %%
# Statistiques de base
stats = con.execute("""
    SELECT 
        COUNT(*) as total_books,
        COUNT(DISTINCT title) as unique_titles,
        COUNT(DISTINCT key) as unique_keys,
        MIN(first_publish_year) as oldest_year,
        MAX(first_publish_year) as newest_year,
        AVG(number_of_pages_median) as avg_pages,
        COUNT(isbn) as books_with_isbn
    FROM library.books
""").fetchdf()

print("📊 Statistiques globales:")
print(stats.T)
```
